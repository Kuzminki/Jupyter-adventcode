```python
f = open("day1-input.txt", "r")
a=0
b=0
GlobalResult=0

for cline in f:
    a=0
    b=0
    indic_a=0
    indic_b=0
    for cchar in cline:
        if cchar.isnumeric():
            if indic_a==0:
                a=cchar
                indic_a=1
            else:
                b=cchar
                indic_b=1
        if indic_b==0:
            b=a
    Resultat=int(a)*10+int(b)
    # print(cline[:-1])
    # print(Resultat)
    GlobalResult+=Resultat
print(GlobalResult)



```

```python
f = open("day1-input.txt", "r")
a=0
b=0
GlobalResult=0

for cline in f:
    a=0
    b=0
    indic_a=0
    indic_b=0
    for index in range(len(cline)):
        if cline[index:index+3]=="one":
            print("! one")
            if indic_a==0:
                a=1
                indic_a=1
            else:
                b=1
                indic_b=1
        elif cline[index:index+3]=='two':
            print("! two")
            if indic_a==0:
                a=2
                indic_a=1
            else:
                b=2
                indic_b=1
        elif cline[index:index+5]=='three':
            print("! three")
            if indic_a==0:
                a=3
                indic_a=1
            else:
                b=3
                indic_b=1
        elif cline[index:index+4]=='four':
            print("! four")
            if indic_a==0:
                a=4
                indic_a=1
            else:
                b=4
                indic_b=1
        elif cline[index:index+4]=='five':
            print("! five")
            if indic_a==0:
                a=5
                indic_a=1
            else:
                b=5
                indic_b=1
        elif cline[index:index+3]=='six':
            print("! six")
            if indic_a==0:
                a=6
                indic_a=1
            else:
                b=6
                indic_b=1
        elif cline[index:index+5]=='seven':
            print("! seven")
            if indic_a==0:
                a=7
                indic_a=1
            else:
                b=7
                indic_b=1
        elif cline[index:index+5]=='eight':
            print("! eight")
            if indic_a==0:
                a=8
                indic_a=1
            else:
                b=8
                indic_b=1
        elif cline[index:index+4]=='nine':
            print("! nine")
            if indic_a==0:
                a=9
                indic_a=1
            else:
                b=9
                indic_b=1
        elif cline[index].isnumeric():
            if indic_a==0:
                a=cline[index]
                indic_a=1
            else:
                b=cline[index]
                indic_b=1
    if indic_b==0:
        b=a
    print(str(a)+" "+str(b))
    GlobalResult+=int(a)*10+int(b)
print(GlobalResult)



```
